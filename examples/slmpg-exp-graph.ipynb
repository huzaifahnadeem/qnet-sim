```python
import os
import pandas as pd
```

```python
csv_dir = './exps'
csv_filenames = os.listdir(csv_dir)
csv_filenames[:] = [x for x in csv_filenames if x != '.DS_Store'] # remove '.DS_Store' if it exists

df_template = {
        'p': [],
        'q': [],
        'seed': [],
        'ts': [],
        'data_qubit_transmitted': [],
        'x-hop': [],
        'y-hop': [],
    }
df = pd.DataFrame(df_template)
```

```python
# e.g 'p0.45-q1-s7.csv'
row_idx = 0
for c in csv_filenames:
    prob_p = float(c.split('-')[0][1:])
    prob_q = float(c.split('-')[1][1:]) 
    seed = int(c.split('-')[2].split('.')[0][1:])
    csv_data = pd.read_csv(f"./{csv_dir}/{c}")
    for i, _ in csv_data.iterrows():
        df_row = []
        for k in df_template.keys():
            if k == 'p':
                df_row.append(prob_p)
            elif k == 'q':
                df_row.append(prob_q)
            elif k == 'seed':
                df_row.append(seed)
            elif k == 'data_qubit_transmitted':
                qubit_status = csv_data[k][i]
                df_row.append(1 if qubit_status == 'success' else 0)
            else:
                df_row.append(csv_data[k][i])
        df.loc[row_idx] = df_row
        row_idx += 1
df
```

```python
convert_dict = {
        'p': float,
        'q': float,
        'seed': int,
        'ts': int,
        'data_qubit_transmitted': int,
        'x-hop': int,
        'y-hop': int,
    }
df = df.astype(convert_dict)
df
```
